```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = '/Users/neharana/Downloads/Multivitamin use â‰¥4 times a week during the month before pregnancy-2016-2021 - Nutrition.csv'  # Replace with your file path
data = pd.read_csv(file_path)
```

```python
# Descriptive Statistics
descriptive_stats = data.describe()

# Trend Analysis - Average Weighted % of multivitamin use over the years
trend_data = data.groupby('Year')['Weighted %'].mean().reset_index()

# Setting the aesthetics for the plots
sns.set(style="whitegrid")

# Plotting the trend
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Weighted %', data=trend_data, marker='o')
plt.title('Trend of Multivitamin Use (Weighted %) from 2016 to 2021')
plt.xlabel('Year')
plt.ylabel('Average Weighted % of Multivitamin Use')
plt.xticks(trend_data['Year'])
plt.show()

# Comparative Analysis - Average Weighted % of multivitamin use across sites
site_data = data.groupby('Site Name')['Weighted %'].mean().sort_values().reset_index()

# Plotting the comparative analysis
plt.figure(figsize=(15, 10))
sns.barplot(x='Weighted %', y='Site Name', data=site_data, palette="viridis")
plt.title('Average Weighted % of Multivitamin Use by Site (2016-2021)')
plt.xlabel('Average Weighted % of Multivitamin Use')
plt.ylabel('Site Name')
plt.show()
```
