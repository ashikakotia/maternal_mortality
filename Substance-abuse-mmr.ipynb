```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is your DataFrame after loading the CSV
data = pd.read_csv('/Users/neharana/Downloads/Substance-Abuse-Pregnancy.csv')
print(data)
```

```python
# Descriptive Statistics
descriptive_stats = data.describe(include='all')
```

```python
cigarette_columns = [col for col in data.columns if 'Cigarette' in col]
cigarette_data = data[cigarette_columns + ['State', 'Year']]

# Displaying the first few rows to understand the structure of these columns
cigarette_data.head()
```

```python
# Descriptive statistics for cigarette smoking metrics
cigarette_descriptive_stats = cigarette_data.describe()
cigarette_descriptive_stats.T
```

```python
# Trend analysis for cigarette smoking rates

# Focusing on the "Weighted %" columns for different stages
trend_columns = [col for col in cigarette_columns if 'Weighted %' in col]
trend_data = data.groupby('Year')[trend_columns].mean()

plt.figure(figsize=(12, 8))

# Plotting trends for each stage
for col in trend_columns:
    plt.plot(trend_data.index, trend_data[col], marker='o', label=col.split(' - ')[0])

plt.title('Trend of Cigarette Smoking Rates During Different Pregnancy Stages')
plt.xlabel('Year')
plt.ylabel('Weighted %')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Focusing on the 'Weighted %' columns for comparison
trend_columns = [col for col in cigarette_data.columns if 'Weighted %' in col]
comparative_data = cigarette_data.groupby('State')[trend_columns].mean()

plt.figure(figsize=(15, 10))
sns.heatmap(comparative_data, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title('Average Cigarette Smoking Rates Across States')
plt.ylabel('State')
plt.xlabel('Pregnancy Stage')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation analysis between smoking rates at different pregnancy stages

correlation_data = cigarette_data[trend_columns]

correlation_matrix = correlation_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Matrix of Cigarette Smoking Rates at Different Pregnancy Stages')
plt.show()
```

The values close to 1.00 indicate very strong positive relationships between the smoking rates at different stages of pregnancy and postpartum. This suggests that those who smoke before or during pregnancy are likely to continue smoking at similar rates postpartum.

```python
# Dropping rows where the relevant smoking data is missing
data_clean_ecig = data.dropna(subset=['Cigarette smoking during the 3 months before pregnancy - Weighted %'])
trend_data_ecig = data_clean_ecig.groupby('Year')['Cigarette smoking during the 3 months before pregnancy - Weighted %'].mean().reset_index()
state_comparison = data_clean_ecig.groupby('State')['Cigarette smoking during the 3 months before pregnancy - Weighted %'].mean().sort_values(ascending=False).reset_index()


```

```python
# State Comparison Plot
plt.plot()
sns.barplot(x='Cigarette smoking during the 3 months before pregnancy - Weighted %', y='State', data=state_comparison.head(10), palette=sns.diverging_palette(20, 220, n=10))
plt.title('Top 10 States with Highest Smoking Rates')
plt.tight_layout()
```

```python
cols_for_correlation = [
    'Cigarette smoking during the 3 months before pregnancy - Weighted %', 
    'Heavy drinking (≥8 drinks a week) during the 3 months before pregnancy -Weighted %', 
    'Deaths/100000'
]

correlation_data = data_clean_ecig[cols_for_correlation].corr()
statistical_summary = data_clean_ecig[cols_for_correlation].describe()

# Correlation Heatmap
plt.subplot()
sns.heatmap(correlation_data, annot=True, cmap='viridis')
plt.title('Correlation Between Variables')

plt.show()
```

```python
The correlation between cigarette smoking before pregnancy and heavy drinking is very low and positive (0.019), suggesting almost no linear relationship.
The correlation between cigarette smoking before pregnancy and deaths per 100,000 is negative (-0.18), which could imply that higher smoking rates are weakly associated with lower death rates, or it might simply be a spurious correlation without causal meaning.
The correlation between heavy drinking before pregnancy and deaths per 100,000 is moderately negative (-0.31), suggesting a weak to moderate inverse relationship.
```

```python
# Selecting the relevant columns for trend analysis
trend_columns = [
    'Year',
    'e-cigarette use during the 3 months before pregnancy - Weighted %',
    'e-cigarette use during the last 3 months of pregnancy - Weighted %'
]

# Filtering the dataset for trend analysis
trend_data = data[trend_columns].dropna()

# Aggregating data by year
trend_data_grouped = trend_data.groupby('Year').mean().reset_index()

# Plotting the trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='e-cigarette use during the 3 months before pregnancy - Weighted %', data=trend_data_grouped, marker='o', label='Before Pregnancy')
sns.lineplot(x='Year', y='e-cigarette use during the last 3 months of pregnancy - Weighted %', data=trend_data_grouped, marker='o', label='During Pregnancy')

plt.title('Trends in E-Cigarette Use Before and During Pregnancy Over Time')
plt.xlabel('Year')
plt.ylabel('Weighted %')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Selecting the most recent year available in the dataset
most_recent_year = trend_data['Year'].max()

# Filtering the dataset for the most recent year
state_comparison_data = data[data['Year'] == most_recent_year][[
    'State',
    'e-cigarette use during the 3 months before pregnancy - Weighted %',
    'e-cigarette use during the last 3 months of pregnancy - Weighted %'
]].dropna()

plt.figure(figsize=(15, 10))

# Plotting 'Before Pregnancy' data in pastel pink
sns.barplot(x='e-cigarette use during the 3 months before pregnancy - Weighted %', y='State', 
            data=state_comparison_data, color='lightpink', label='Before Pregnancy')

# Plotting 'During Pregnancy' data in pastel purple
sns.barplot(x='e-cigarette use during the last 3 months of pregnancy - Weighted %', y='State', 
            data=state_comparison_data, color='plum', label='During Pregnancy')

plt.title(f'Comparison of E-Cigarette Use by State in {most_recent_year}')
plt.xlabel('Weighted %')
plt.ylabel('State')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Check the range of years in the dataset
years_range = data['Year'].unique()

# Calculate the average hookah use per year
average_hookah_per_year = data.groupby('Year')['Hookah use in the last 2 years - Weighted %'].mean().reset_index()

# Plotting the trend over years
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Hookah use in the last 2 years - Weighted %', data=average_hookah_per_year, marker='o')
plt.title('Trend of Hookah Use Over Years')
plt.xlabel('Year')
plt.ylabel('Average Weighted % of Hookah Use')
plt.grid(True)
plt.show()

```

```python
average_hookah_per_state = data.groupby('State')['Hookah use in the last 2 years - Weighted %'].mean().reset_index()

# Sorting to find top and bottom states
top_states_hookah = average_hookah_per_state.sort_values(by='Hookah use in the last 2 years - Weighted %', ascending=False).head(5)
bottom_states_hookah = average_hookah_per_state.sort_values(by='Hookah use in the last 2 years - Weighted %', ascending=True).head(5)

top_states_hookah, bottom_states_hookah
# Selecting a few states for detailed trend analysis
selected_states = list(top_states_hookah['State']) + list(bottom_states_hookah['State'])

# Filtering the data for the selected states
selected_states_data = data[data['State'].isin(selected_states)]

# Plotting the trends
plt.figure(figsize=(15, 10))

sns.lineplot(x='Year', y='Hookah use in the last 2 years - Weighted %', hue='State', 
             data=selected_states_data, marker='o')

plt.title('Yearly Trends of Hookah Use in Selected States')
plt.xlabel('Year')
plt.ylabel('Weighted % of Hookah Use')
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
